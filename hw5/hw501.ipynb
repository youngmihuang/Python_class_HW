- 自訂函數 sigmoid_fun() 並且作圖（自由取一段 x 區間）

承前題，將前兩個函數畫在同一個函數圖（Optional）；
承前題，利用 subplot() 將這兩個圖畫在 (1, 2) 的畫布上（Optional）

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 建立data: sign_fun()-another way
x = np.linspace(-10, 10, 1000)
y = np.sign(x)
plt.plot(x, y, color='blue', linewidth=2.5, linestyle='-', label= '$sign_fun')
# Moving spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.yticks([-1, 0, 1], ['$-1$', '$0$', '$+1$'])

#新增圖例
plt.legend(loc='upper left')
plt.show()
```

```python
# 建立data: sigmoid_fun()
import math
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
x = range(-10,11)
y = sigmoid(x)


#畫圖
plt.plot(x,y, color='red', linewidth=2.5, linestyle='-', label= '$sigmoid_fun')
# Moving spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.yticks([ 0, 1], ['$0$', '$+1$'])

#新增圖例
plt.legend(loc='upper left')
plt.show()
```

```python
#Optional 建在同一張圖
#建立data
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
x = np.linspace(-10, 10, 1000)
sign_fun, sigmoid_fun = np.sign(x), sigmoid(x)

#畫圖
plt.plot(x,sign_fun, color='blue', linewidth=2.5, linestyle='-', label= 'sign_fun')
plt.plot(x,sigmoid_fun, color='red', linewidth=2.5, linestyle='-', label= 'sigmoid_fun')

# Moving spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.yticks([ -1, 0, 1], [ '$-1$','$0$', '$+1$'])


#新增圖例_移除邊框
plt.legend(loc='upper left', frameon = False)  
plt.show()
```

```python
#Optional 建subplot
#建立data
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
x = np.linspace(-10, 10, 1000)
sign_fun, sigmoid_fun = np.sign(x), sigmoid(x)


#畫圖
plt.figure(figsize=(5,5), dpi = 80)
plt.subplot(2,1,1)
plt.plot(x,sign_fun, color='blue', linewidth=2.5, linestyle='-', label= 'sign_fun')
plt.legend(loc = 'upper left')

plt.subplot(2,1,2)
plt.plot(x,sigmoid_fun, color='red', linewidth=2.5, linestyle='-', label= 'sigmoid_fun')


#新增圖例
plt.legend(loc = 'upper left') 
plt.show()
```
