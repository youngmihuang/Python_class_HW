- 練習將 https://storage.googleapis.com/py_ml_datasets/train.csv 中的 Sex 與 Embarked 由類別型變數轉換為 One-hot encoding 的 dummy variables

```python
#import url: data
import pandas as pd
import numpy as np
import sklearn

url = 'https://storage.googleapis.com/py_ml_datasets/train.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
#Convert Sex Start
data_df = data.groupby(['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']).size().reset_index(name='count')
data_df.head()
```

```python
pivot_df = data_df.pivot_table(index=['PassengerId','Survived','Pclass','Name','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'], columns='Sex', values='count').fillna(0)
pivot_df = pivot_df.reindex_axis(['male', 'female'], axis=1)
pivot_df.head()
```

```python
turn_1 = pivot_df.reset_index(['PassengerId','Survived','Pclass','Name','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'])
```

```python
turn_1.head()
```

```python
#Convert Sex finish
turn_2 = pd.DataFrame({'PassengerId': turn_1.PassengerId,'Survived': turn_1.Survived, 'Pclass': turn_1.Pclass ,'Name': turn_1.Name,'Age': turn_1.Age,'SibSp': turn_1.SibSp,'Parch': turn_1.Parch,'Ticket': turn_1.Ticket,'Fare': turn_1.Fare,'Cabin': turn_1.Cabin,'Embarked':turn_1.Embarked,'Male': turn_1.male,'Female': turn_1.female})
turn_2.head()
```

```python
#Convert Embarked start
embarked = turn_2.groupby(['PassengerId','Survived','Pclass','Name','Female','Male','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']).size().reset_index(name='count')
embarked.head()
```

```python
embark_df = embarked.pivot_table(index=['PassengerId','Survived','Female','Male','Pclass','Name','Age','SibSp','Parch','Ticket','Fare','Cabin'], columns='Embarked', values='count').fillna(0)
embark_df = embark_df.reindex_axis(['C', 'S','Q'], axis=1)
embark_df.head()
```

```python
turn_embark = embark_df.reset_index(['PassengerId','Survived','Female','Male','Pclass','Name','Age','SibSp','Parch','Ticket','Fare','Cabin'])
turn_embark.head()
```

```python
#Convert Embarked finish
#Final output
turn_final = pd.DataFrame({'PassengerId': turn_embark.PassengerId,'Survived': turn_embark.Survived, 'Pclass': turn_embark.Pclass ,'Name': turn_embark.Name,'Age': turn_embark.Age,'SibSp': turn_embark.SibSp,'Parch': turn_embark.Parch,'Ticket': turn_embark.Ticket,'Fare': turn_embark.Fare,'Cabin': turn_embark.Cabin,'Male': turn_embark.Male,'Female': turn_embark.Female,'Embarked_C': turn_embark.C,'Embarked_S': turn_embark.S,'Embarked_Q': turn_embark.Q})
turn_final.head()
```
